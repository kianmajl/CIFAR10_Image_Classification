### Dense (fully connected) layer

```python
class Dense:
    def __init__(self, n_inputs, n_neurons):
        self.n_inputs = n_inputs
        self.n_neurons = n_neurons

        # Initialize the weights with random values based on normal distribution with He heuristic
        self.weights = (
            torch.randn(n_inputs, n_neurons) * torch.sqrt(torch.tensor(2.0 / n_inputs))
        ).to(device)
        # Initialize the biases with zeros
        self.biases = torch.zeros(1, n_neurons).to(device)

    def forward(self, inputs):
        self.inputs = inputs
        self.output = torch.matmul(self.inputs, self.weights) + self.biases

    def backward(self, b_input):
        self.weights_gradient = torch.matmul(self.inputs.T, b_input)
        self.biases_gradient = torch.sum(b_input, dim=0, keepdims=True)
        self.inputs_gradient = torch.matmul(b_input, self.weights.T)

        self.b_output = self.inputs_gradient
```

### Activation functions

```python
class ReLU:
    def __repr__(self):
        return "ReLU"

    def forward(self, inputs):
        # Calculate the output using the ReLU activation function
        self.output = torch.maximum(torch.zeros(inputs.size()).to(device), inputs)

    def backward(self, b_input):
        inputs_gradient = b_input.clone()
        # Set the gradient to zero where the output is less than or equal to zero
        inputs_gradient[self.output <= 0] = 0
        self.b_output = inputs_gradient
```

```python
class Sigmoid:
    def __repr__(self):
        return "Sigmoid"

    def forward(self, inputs):
        # Calculate the sigmoid function of the input tensor
        self.output = 1 / (1 + torch.exp(-inputs))

    def backward(self, b_input):
        inputs_gradient = b_input * (1 - self.output) * self.output
        self.b_output = inputs_gradient
```

### Loss function

```python
class Categorical_Cross_Entropy_loss_Softmax:
    def forward(self, inputs, batch_size, class_label=None):
        max_out, max_inds = torch.max(inputs, dim=1, keepdims=True)

        # Calculate the exponential values
        exp_vals = torch.exp(inputs - max_out)

        # Calculate the softmax values
        self.softmax = exp_vals / torch.sum(exp_vals, dim=1, keepdims=True)
        self.softmax += 1e-7  # Add a small value to avoid division by zero

        # Calculate the loss
        if class_label is not None:
            self.class_label = class_label
            self.batch_size = batch_size
            self.loss = -torch.sum(self.class_label * torch.log(self.softmax))

    def backward(self):
        # Calculate the gradient of the inputs
        self.b_output = (self.softmax - self.class_label) / self.batch_size
```

### Optimizer

```python
class SGD:
    def __init__(self, learning_rate=0.001):
        self.learning_rate = learning_rate

    def update(self, layer):
        layer.weights -= (
            self.learning_rate * layer.weights_gradient
        )  # Update the weights
        layer.biases -= self.learning_rate * layer.biases_gradient  # Update the biases
```

### Architecture

#### Features Extractor

```python
class ResNet34:
    def __init__(self):
        # Load the ResNet34 model with default weights
        self.resnet34 = resnet34(weights=ResNet34_Weights.DEFAULT)
        modules = list(self.resnet34.children())[:-1]

        # Create a new model with all the layers except the last one
        self.resnet34 = torch.nn.Sequential(*modules)

        self.resnet34.eval()  # Set the model to evaluation mode

        self.resnet34 = self.resnet34.to(device)

    def __repr__(self):
        return "ResNet34"

    def get_features(self, images):
        with torch.no_grad():
            # Get the features from the features extractor model
            features = self.resnet34(images)
        return features

    def get_size(self):
        with torch.no_grad():
            features = torch.flatten(
                self.resnet34(torch.zeros(1, 3, 224, 224).to(device)), start_dim=1
            )
        # Return the size of the features
        return features.shape
```

```python
fe_resnet18 = ResNet18()
fe_resnet34 = ResNet34()
fe_vgg11 = VGG11()
```

#### Neural Network

```python
class AKModel:
    def __init__(
        self,
        features_extractor=fe_resnet34,
        number_of_classes=10,
        batch_size=32,
        tune=False,
    ):
        self.batch_size = batch_size
        self.number_of_classes = number_of_classes

        self.optimizer = SGD(learning_rate=0.001)
        self.features_extractor = features_extractor
        self.n_inputs = self.features_extractor.get_size()[1]

        if not tune:
            self.layers = [Dense(self.n_inputs, 20), Dense(20, self.number_of_classes)]
            self.activations = [ReLU(), Categorical_Cross_Entropy_loss_Softmax()]

    def _forward_propagation(self, data, y_1hot=None):
        """
        This function performs forward propagation on a neural network.
        It takes as input a data tensor and an one-hot encoded label tensor.
        """
        if len(self.activations) != len(self.layers):
            raise Exception(
                "The number of activations should be equal to the number of layers"
            )

        inputs = torch.flatten(self.features_extractor.get_features(data), start_dim=1)
        for layer, activation in zip(self.layers, self.activations):
            layer.forward(inputs)
            if activation == self.activations[-1]:
                activation.forward(layer.output, self.batch_size, y_1hot)
                break
            else:
                activation.forward(layer.output)
            inputs = activation.output

    def _test_val(self):
        acc_hist = []
        loss_hist = []
        for data, label in self.val_loader:
            data = data.to(device)
            label = label.to(device)
            y_1hot = torch.nn.functional.one_hot(
                label, num_classes=self.number_of_classes
            )

            # Forward propagation
            self._forward_propagation(data, y_1hot)

            # Calculate the accuracy
            y_predict = torch.argmax(self.activations[-1].softmax, dim=1)
            accuracy = torch.mean((torch.argmax(y_1hot, dim=1) == y_predict).float())
            acc_hist.append(accuracy)

            # Calculate the loss
            loss = self.activations[-1].loss
            loss_hist.append(loss)

        return torch.mean(torch.Tensor(loss_hist)), torch.mean(torch.Tensor(acc_hist))

    def evaluate(self, train=False):
        if train:
            y_true = []
            y_predict = []

            for data, label in self.train_loader:
                data = data.to(device)
                label = label.to(device)
                y_1hot = torch.nn.functional.one_hot(
                    label, num_classes=self.number_of_classes
                )

                # Forward propagation
                self._forward_propagation(data, y_1hot)

                y_pred = torch.argmax(self.activations[-1].softmax, dim=1)

                for yt, yp in zip(torch.argmax(y_1hot, dim=1), y_pred):
                    y_true.append(yt.item())
                    y_predict.append(yp.item())

            return (
                accuracy_score(y_true, y_predict),
                f1_score(y_true, y_predict, average="weighted"),
            )

        else:
            y_true = []
            y_predict = []

            for data, label in self.test_loader:
                data = data.to(device)
                label = label.to(device)
                y_1hot = torch.nn.functional.one_hot(
                    label, num_classes=self.number_of_classes
                )

                # Forward propagation
                self._forward_propagation(data, y_1hot)

                y_pred = torch.argmax(self.activations[-1].softmax, dim=1)

                for yt, yp in zip(torch.argmax(y_1hot, dim=1), y_pred):
                    y_true.append(yt.item())
                    y_predict.append(yp.item())

            return (
                accuracy_score(y_true, y_predict),
                f1_score(y_true, y_predict, average="weighted"),
            )

    def predict(self, data):
        data = data.to(device)

        # Forward propagation
        self._forward_propagation(data)

        y_predict = torch.argmax(self.activations[-1].softmax, dim=1)
        return y_predict

    def train(self, epochs=20, disable_tqdm=False):
        self.epochs = epochs

        self.train_loss_history = []
        self.val_loss_history = []

        self.train_acc_history = []
        self.val_acc_history = []

        for epoch in range(epochs):
            with tqdm(self.train_loader, unit="batch", disable=disable_tqdm) as tepoch:
                # Initialize the history lists for each epoch
                acc_hist = []
                loss_hist = []
                for data, label in tepoch:
                    tepoch.set_description(f"Epoch {epoch + 1}")

                    # Transfer the data to the GPU
                    data = data.to(device)
                    label = label.to(device)

                    # Convert the labels to one-hot encoded vectors
                    y_1hot = torch.nn.functional.one_hot(
                        label, num_classes=self.number_of_classes
                    )

                    self._forward_propagation(data, y_1hot)

                    # Calculate the accuracy
                    y_predict = torch.argmax(self.activations[-1].softmax, dim=1)
                    accuracy = torch.mean(
                        (torch.argmax(y_1hot, dim=1) == y_predict).float()
                    )
                    acc_hist.append(accuracy)

                    # Calculate the loss
                    loss = self.activations[-1].loss
                    loss_hist.append(loss)

                    # Backward propagation
                    self.activations[-1].backward()
                    b_input = self.activations[-1].b_output
                    for layer, activation in zip(
                        reversed(self.layers), reversed(self.activations[:-1])
                    ):
                        layer.backward(b_input)
                        activation.backward(layer.b_output)
                        b_input = activation.b_output
                    self.layers[0].backward(b_input)

                    # Update the weights and biases
                    for layer in self.layers:
                        self.optimizer.update(layer)

                    tepoch.set_postfix(
                        loss=loss.item(), accuracy=f"{accuracy.item()*100:.2f}%"
                    )

            # Calculate the validation loss and accuracy
            val_loss, val_acc = self._test_val()

            # Append the history lists
            self.train_loss_history.append(torch.mean(torch.Tensor(loss_hist)))
            self.val_loss_history.append(val_loss)

            self.train_acc_history.append(torch.mean(torch.Tensor(acc_hist)) * 100)
            self.val_acc_history.append(val_acc * 100)

            # Print the results
            if not disable_tqdm:
                print(f"Epoch:{epoch+1}")
                print(f"Training Loss: {self.train_loss_history[-1]}")
                print(f"Validation Loss: {self.val_loss_history[-1]}")
                print(f"Training Accuracy: {self.train_acc_history[-1]:.2f}%")
                print(f"Validation Accuracy: {self.val_acc_history[-1]:.2f}%")
                print("--------------------------------------------------")

    def plot_losses_and_accuracy(self):
        x_range = range(1, self.epochs + 1)

        plt.figure(figsize=(15, 10))
        plt.xticks(x_range)
        plt.plot(
            x_range, self.train_acc_history, label="Training Accuracy", color="blue"
        )
        plt.plot(
            x_range, self.val_acc_history, label="Validation Accuracy", color="red"
        )
        plt.legend()
        plt.ylabel("Accuracy")
        plt.xlabel("Epoch")
        plt.gca().xaxis.set_major_locator(MultipleLocator(1))
        plt.xlim([1, self.epochs])
        plt.savefig("Accuracy.png")
        plt.show()

        plt.figure(figsize=(15, 10))
        plt.xticks(x_range)
        plt.plot(x_range, self.train_loss_history, label="Training Loss", color="blue")
        plt.plot(x_range, self.val_loss_history, label="Validation Loss", color="red")
        plt.legend()
        plt.ylabel("Loss")
        plt.xlabel("Epoch")
        plt.gca().xaxis.set_major_locator(MultipleLocator(1))
        plt.xlim([1, self.epochs])
        plt.savefig("Loss.png")
        plt.show()

    
```

### NAS

```python
MAX_EPOCHS = 5
POPULATION_SIZE = 10
MAX_GENERATION = 10
MAX_EXE = 5
```

```python
class Chromosome:
    def __init__(self):
        self.features_extractor_network = None
        self.number_of_hidden_layers = None
        self.learning_rate = None
        self.batch_size = None
        self.fitness = None
        self.number_of_neurons_per_layer = []
        self.activation_functions_per_layer = []

    def __repr__(self):
        return (
            f"features_extractor_network: {self.features_extractor_network}\n"
            f"number_of_hidden_layers: {self.number_of_hidden_layers}\n"
            f"learning_rate: {self.learning_rate}\n"
            f"batch_size: {self.batch_size}\n"
            f"fitness: {self.fitness}\n"
            f"number_of_neurons_per_layer: {self.number_of_neurons_per_layer}\n"
            f"activation_functions_per_layer: {self.activation_functions_per_layer}\n"
        )
    
    def mutate(self):
        # Randomly select a new features extractor network
        self.features_extractor_network = np.random.choice(
            [fe_resnet18, fe_resnet34, fe_vgg11]
        )

        # Mutation of the number of hidden layers
        self.number_of_hidden_layers = np.random.choice([0, 1, 2])

        # Increase/Decrease learning rate by a factor of three with probability 0.5
        if np.random.rand() <= 0.5:
            self.learning_rate *= 3
        else:
            self.learning_rate /= 3

        # Double/Halve batch size with probability 0.5
        if np.random.rand() <= 0.5:
            self.batch_size *= 2
        else:
            self.batch_size /= 2

        self.batch_size = max(1, self.batch_size)  # Ensure batch size is at least one

        # Randomly select a new number of neurons for each hidden layer
        self.number_of_neurons_per_layer.clear()
        for i in range(self.number_of_hidden_layers):
            self.number_of_neurons_per_layer.append(np.random.choice([10, 20, 30]))

        # Add ReLU/Sigmoid activation function with probability 0.5
        self.activation_functions_per_layer.clear()
        for i in range(self.number_of_hidden_layers):
            if np.random.rand() <= 0.5:
                self.activation_functions_per_layer.append(ReLU())
            else:
                self.activation_functions_per_layer.append(Sigmoid())

    @staticmethod
    def recombinate(parent_1, parent_2):
        offspring1, offspring2 = Chromosome(), Chromosome()

        # Recombination of the features extractor network
        offspring1.features_extractor_network = parent_1.features_extractor_network
        offspring2.features_extractor_network = parent_2.features_extractor_network

        # Recombination of the number of hidden layers
        offspring1.number_of_hidden_layers = parent_1.number_of_hidden_layers
        offspring2.number_of_hidden_layers = parent_2.number_of_hidden_layers

        # Combine learning rates using a random recombination factor
        alpha1 = np.random.rand()
        offspring1.learning_rate = (alpha1 * parent_1.learning_rate) + (
            (1 - alpha1) * parent_2.learning_rate
        )

        alpha2 = np.random.rand()
        offspring2.learning_rate = (alpha2 * parent_1.learning_rate) + (
            (1 - alpha2) * parent_2.learning_rate
        )

        # Recombination of the batch size
        offspring1.batch_size = parent_1.batch_size
        offspring2.batch_size = parent_2.batch_size

        # Recombination of the number of neurons per layer
        parents_number_of_neurons_per_layer = (
            parent_1.number_of_neurons_per_layer + parent_2.number_of_neurons_per_layer
        )

        offspring1.number_of_neurons_per_layer = np.random.choice(
            parents_number_of_neurons_per_layer, offspring1.number_of_hidden_layers
        ).tolist()
        offspring2.number_of_neurons_per_layer = np.random.choice(
            parents_number_of_neurons_per_layer, offspring2.number_of_hidden_layers
        ).tolist()

        # Recombination of the activation functions per layer
        parents_activation_functions_per_layer = deepcopy(
            parent_1.activation_functions_per_layer
        ) + deepcopy(parent_2.activation_functions_per_layer)

        offspring1.activation_functions_per_layer = np.random.choice(
            parents_activation_functions_per_layer, offspring1.number_of_hidden_layers
        ).tolist()
        offspring2.activation_functions_per_layer = np.random.choice(
            parents_activation_functions_per_layer, offspring2.number_of_hidden_layers
        ).tolist()

        return offspring1, offspring2

    def build_model(self):
        self.model = AKModel(
            features_extractor=self.features_extractor_network,
            batch_size=int(self.batch_size.item()),
            tune=True,
        )

        self.model.train_loader = torch.utils.data.DataLoader(
            dataset=train_subset_nas,
            shuffle=True,
            batch_size=int(self.batch_size.item()),
        )
        self.model.val_loader = torch.utils.data.DataLoader(
            dataset=val_subset_nas,
            shuffle=False,
            batch_size=int(self.batch_size.item()),
        )
        self.model.test_loader = torch.utils.data.DataLoader(
            dataset=test_data_nas, shuffle=False, batch_size=int(self.batch_size.item())
        )
        self.model.optimizer.learning_rate = self.learning_rate.item()
        self.model.n_inputs = 512
        self.model.activations = deepcopy(self.activation_functions_per_layer)
        self.model.layers = []

        self.model.activations.append(Categorical_Cross_Entropy_loss_Softmax())

        for i in range(self.number_of_hidden_layers):
            if i == 0:
                self.model.layers.append(
                    Dense(self.model.n_inputs, self.number_of_neurons_per_layer[i])
                )
            else:
                self.model.layers.append(
                    Dense(
                        self.number_of_neurons_per_layer[i - 1],
                        self.number_of_neurons_per_layer[i],
                    )
                )

        if self.number_of_hidden_layers:
            self.model.layers.append(
                Dense(
                    self.number_of_neurons_per_layer[-1], self.model.number_of_classes
                )
            )
        else:
            self.model.layers.append(
                Dense(self.model.n_inputs, self.model.number_of_classes)
            )

    def calculate_fitness(self):
        acc_hist = []

        for __ in range(MAX_EXE):
            self.build_model()
            self.model.train(epochs=MAX_EPOCHS, disable_tqdm=True)
            acc_hist.append(self.model.evaluate(train=False)[0])

        self.fitness = np.mean(acc_hist)
        del self.model
        return self.fitness
```
